Basic List Programs


List is enclosed in square bracket[]

Let's define one list and check its datatype. Then we'll go on exploring different properties and functions of a list.

```python
# define a list
courses = ['ba','bcom','bsc','ma','mcom','msc']
print(courses) # Note that the output is in square bracket. Remember, square brackets denote list
```

```python
# It's a good practice to know data type of a variable, so we know what we are dealing with

print(type(courses))
```

Accessing List items


```python
# Access first element
courses = ['ba','bcom','bsc','ma','mcom','msc']
print(courses[0])
```

```python
# Access third element

print(courses[2])
```

```python
# Access first and second element

print(courses[0:2])
```

```python
# Access 2nd to 5th element (i.e bcom bsc ma mcom)
# 2nd element = 1st index, 5th element = 4th index

print(courses[1:5])   # this is nothing but courses[1], courses[2], courses[3], courses[3]
# when using ranges, the output EXCLUDES the second number (hence, if we write index 5, then we get index 4th as max)
```

List functions

```python
# Let's first check what all functions list supports
# ignore the functions that start with __

help(list)       # list is a special and reserved keyword in python. It should NOT be used as variablename
```

If you run the above function twice, you'll get duplicate entry of BE. Meaning, list allows duplicates

Note - append() always appends at the end of the list

```python
courses = ['ba','bcom','bsc','ma','mcom','msc']
courses.insert(3,'ME') # # i.e. insert requires position and value
print(courses)
```

```python
# You can append another variable as well

c = 'B.Pharm'
courses.append(c)
print(courses)
```

```python
# You can append a list within a list

new = ['11th', '12th']   # Defining a new list
courses.append(new)      # Appending new list to the old list
print(courses)           # Note that the list itself got inserted into the list. That list constitutes one element and not two.
```

```python
print(courses[-1])
```

```python
# Note, in the above example, the list got inserted as a list only and not as individual items.
# To add individual list items to the original list, use the function extend

help(courses.extend)
```

```python
# The last 4 words are important in above definition - ELEMENTS FROM THE ITERABLE. ITERABLE in this case is thew new list that we are passing (see the first argument being passed to the function definiion)
# In other words, extend actually unpacks the other list and adds items to the original list instead of creating a nested list by simply inserting one list into the other

# Let's check
courses.extend(new)
print(courses)

```

```python
# To deconstruct what is exactly happening in the above code, you can use print() like this

cubes = []
for i in range(5):
  print("i = ", i)
  cubes.append(i ** 3)
  print("Number appended (i ** 3) = ", i ** 3)
  print("Current status of cubes list = ", cubes)
print()
print('Outside for loop')
print(cubes)
```
